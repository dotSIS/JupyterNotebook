# 1) Remove Punctuation

```python
def remove_punct(my_str):
    # define punctuation
    punctuations = '''!()-[]{};:'.'"\,<>./?@#$%^-&_0123456789~*''' #â€™
    
    # remove punctuation from the string
    no_punct = ""
    for char in my_str:
        if char not in punctuations:
            no_punct = no_punct + char
    return no_punct
```

# 2) Removing Repeated Letters

```python
from nltk.corpus import wordnet

def removeRepeatedCharacters(tokens):
    repeatPattern = re.compile(r'(\w*)(\w)\2(\w*)')
    matchSubstitution = r'\1\2\3'
    def replace(oldWord):
        if wordnet.synsets(oldWord):
            return(oldWord)
        newWord = repeatPattern.sub(matchSubstitution, oldWord)
        return replace(newWord) if newWord!= oldWord else newWord
    correctTokens = [replace(word) for word in tokens]
    return correctTokens

sentence = ' My Schoooooool is reeeeeeaaaallllllllly amaaaaaazingggg!'
sampleSentence = tokenizeText2(sentence)
print("Original Sentence ", sampleSentence)
print("Corrected Sentence ", removeRepeatedCharacters(sampleSentence))
```

```python
def expand_contractions(text, contraction_mapping=CONTRACTION_MAP):
    
    contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())), 
                                      flags=re.IGNORECASE|re.DOTALL)
#     print(contractions_pattern)
    def expand_match(contraction):
#         print(contraction)
        match = contraction.group(0)
#         print(match)
        first_char = match[0]
#         print(first_char)
        expanded_contraction = contraction_mapping.get(match)\
                                if contraction_mapping.get(match)\
                                else contraction_mapping.get(match.lower())   
#         print(expanded_contraction)
        expanded_contraction = first_char+expanded_contraction[1:]
        return expanded_contraction
        
    expanded_text = contractions_pattern.sub(expand_match, text)
    expanded_text = re.sub("'", "", expanded_text)
    return expanded_text
```

# 4) stopwords removal

```python
from nltk.corpus import stopwords
```

```python
text = """In accordance with existing laws and regulations, and with the assistance 
of the campus office for Research and Innovation, reproduction for 
commercialization processes shall commence, 
including copyright request from the national library, licensing agreements, 
or conditions/agreements between publisher and proponent."""
```

```python
myWordDictionary = []
def remove_stopwords_orig(text):
    data = text
    stopWords = set(stopwords.words('english'))
    words = data
    
    wordsFiltered = []
    for w in words:
        if w not in stopWords:
            myWordDictionary.append(w)
            wordsFiltered.append(w)
            wordsFiltered.append('')
    print('Inside remove stop words: ',wordsFiltered)  
    return "".join(wordsFiltered)
```

```python
for y in myCorpus:
    y = y.lower()
    expandedText = expand_contractions(y)
    remPunct = remove_punct(expandedText)
    tokenList  = tokenizeText2(remPunct)    
    print(tokenList)
```
