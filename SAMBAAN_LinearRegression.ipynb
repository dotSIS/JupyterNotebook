```python
# Import required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics, preprocessing
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, make_scorer

from sklearn.linear_model import LinearRegression
```

# Task 1

```python
# Load the above dataset locally
dt = pd.read_csv('data.csv')
```

```python
# Count the number of missing values
dt.isna().sum()
```

```python
# Replace the missing values with the average of non-null values
# This function cleans the dataset
def clean(data):
  cols = ['RM']
  for col in cols:
    data[col].fillna(data[col].median(), inplace=True)
  return data
```

```python
data = clean(dt)
```

```python
data.isna().sum()
```

```python
# Perform data normalization to all the features
data_scaler = MinMaxScaler()
data_scaled = data_scaler.fit_transform(data)
data_scaled
```

```python
# Perform data standardization to all the features
scaler_standard = StandardScaler()
scaled_std = scaler_standard.fit_transform(data_scaled)
scaled_std
```

```python
# Convert numpy array to pandas dataframe
data = pd.DataFrame(scaled_std, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
```

#Task 2

```python
# Establish your X and Y matrices based on the standardized values
x = data.drop(['CRIM'], axis=1)
y = data[['CRIM']]
x
```

#Task 3

```python
# Perform 75/25 Data split for your X and Y matrix
trainX, testX, trainY, testY = train_test_split(x, y, test_size=0.25)
```

#Task 4

```python
# Provide data dimension for train data test data
print("trainX dimension: ", trainX.shape)
print("trainY dimension: ", trainY.shape)
print("testX dimension: ", testX.shape)
print("testY dimension: ", testY.shape)
```

# Task 5

```python
# Define the linear regression model
lr = LinearRegression()
```

#Task 6

```python
# Build training model
lr.fit(trainX, trainY.values.ravel())
```

#Task 7

```python
# Perform prediction on test data
y_predicted = lr.predict(testX)
```

#Task 8

```python
# Print model performance
print("Model score: {:.2f}%".format(lr.score(testX, testY)))
```
