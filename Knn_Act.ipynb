```python
# Import required modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load fisheriris file and draw dataframe
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= ['Sepal_Length_Cm','Sepal_Width_Cm','Petal_Length_Cm','Petal_Width_Cm','Species'])
```

```python
# Make x and y variable and split dataset
x = df.drop(labels=['Species'],axis=1).values
y = df['Species'].values
trainX, testX, trainY, testY = train_test_split(x, y, test_size=0.2)
```

```python
# Create Knn model with nearest neighbors set to 12
knn = KNeighborsClassifier(n_neighbors=15)
# Train the model
knn.fit(trainX, trainY)
# Predict
y_predicted = knn.predict(testX)
```

```python
# Test DataFrame
df_test=pd.DataFrame(testX, columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
df_test['Actual'] = testY
pred = knn.predict(testX)
df_test['Predicted'] = pred
```

```python
sns.lmplot(x="PetalLengthCm", y="PetalWidthCm", hue='Actual', data=df_test, fit_reg=False, legend=False)
plt.legend(title='LEGEND', loc='lower right', labels=['setosa', 'versicolor', 'virginica'])
plt.show()
```

```python
sns.lmplot(x="PetalLengthCm", y="PetalWidthCm", hue='Predicted', data=df_test, fit_reg=False, legend=False)
plt.legend(title='LEGEND', loc='lower right', labels=['setosa', 'versicolor', 'virginica'])
plt.show()
```
