```python
# File loading above is obsolete and sould be removed
# This will filter missing values manually and load the dataset to a new variable
missing_value = ['#NUM!', np.nan]
df = pd.read_csv('data-water.csv', na_values = missing_value)
```

```python
# Check for missing values
df.isna().sum()
```

```python
# Drop entries that have missing values
df.dropna(subset=['ammonia', 'is_safe'], axis=0, inplace=True)
```

```python
# Check for missing values again
df.isna().sum()
```

#Task 2

```python
# Establish your X and Y matrices
x = df.drop(['is_safe'], axis=1)
y = df[['is_safe']]
```

#Task 3

```python
# Perform 80/20 Data split for your X and Y matrix
trainX, testX, trainY, testY = train_test_split(x, y, test_size=0.20)
```

#Task 4

```python
# Provide data dimension for
# train data
# test data
print("trainX dimension: ", trainX.shape)
print("trainY dimension: ", trainY.shape)
print("testX dimension: ", testX.shape)
print("testY dimension: ", testY.shape)
```

# Task 5

```python
# Define, build, predict the following classification learning algorithms
# Logistic Regression
# Classification Tree
# Random Forest
# Naive-Bayes Classification
lrc = LogisticRegression(max_iter=10000) #10000 cause dataset have 8000 entries
lrc.fit(trainX, trainY.values.ravel())

dtc = DecisionTreeClassifier()
dtc.fit(trainX, trainY.values.ravel())

rfc = RandomForestClassifier()
rfc.fit(trainX, trainY.values.ravel())

nbc = GaussianNB()
nbc.fit(trainX, trainY.values.ravel())
```

```python
lrc_Ypredicted = lrc.predict(testX)
dtc_Ypredicted = dtc.predict(testX)
rfc_Ypredicted = rfc.predict(testX)
nbc_Ypredicted = nbc.predict(testX)
```

#Task 6

```python
# Print the performance of the above models in training and testing data
print("TRAIN MODEL SCORES:")
print("LRC: {:.2f}%".format(lrc.score(trainX, trainY)))
print("DTC: {:.2f}%".format(dtc.score(trainX, trainY)))
print("RFC: {:.2f}%".format(rfc.score(trainX, trainY)))
print("NBC: {:.2f}%".format(nbc.score(trainX, trainY)))
print("TEST MODEL SCORES:")
print("LRC: {:.2f}%".format(lrc.score(testX, testY)))
print("DTC: {:.2f}%".format(dtc.score(testX, testY)))
print("RFC: {:.2f}%".format(rfc.score(testX, testY)))
print("NBC: {:.2f}%".format(nbc.score(testX, testY)))
```
