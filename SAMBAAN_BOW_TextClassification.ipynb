## 3) Display the number of rows and columns for the Bag-of-Words extracted feature from the dataset

```python
def bow_extractor(corpus, ngram_range=(1,1)):
  vectorizer = CountVectorizer(min_df=1, ngram_range=ngram_range)
  features = vectorizer.fit_transform(corpus)
  return vectorizer, features
```

```python
bow_vectorizer, bow_features = bow_extractor(amazon_df['review'])
```

```python
features = bow_features.todense()
print(features) 
```

```python
# feature_names = bow_vectorizer.get_feature_names()
feature_names = bow_vectorizer.get_feature_names_out()
print(feature_names)
```

## 5) Perform text classification for Bag-of-Words feature extraction using Logistic Regression

### -Define x and y matrix. x as the sentences, y as the label

```python
x = amazon_df['review'].values
y = amazon_df['sentiment'].values
```

### -Perform train-test-split using a test size of 0.25, random state of 101

```python
sentences_train, sentences_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=101)
```

### -Display accuracy, precision, recall and F1 score

```python
vectorizer = CountVectorizer()
vectorizer.fit(sentences_train)

X_train = vectorizer.transform(sentences_train)
X_test = vectorizer.transform(sentences_test)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_predicted = classifier.predict(X_test)
```

```python
conf = confusion_matrix(y_test, y_predicted)
```

```python
tp = conf[0][0]
fn = conf[0][1]
fp = conf[1][0]
tn = conf[1][1]
```

```python
precision = tp/(tp+fp)
accuracy = (tp+tn)/(tp+fn+fp+tn)
recall = tp/(tp+fn)
f1 = 2*(precision*recall)/(precision+recall)
lrpre, lracc, lrrec, lrf1 = precision, accuracy, recall, f1
print("Precision: ", lrpre)
print("Accuracy: ", lracc)
print("Recall: ", lrrec)
print("F1-Score: ", lrf1)
```

### -Display Confusion Matrix

```python
results = [("LR", lrpre, lracc, lrrec, lrf1)]
          
pd.options.display.float_format = '{:,.2f}'.format
dtable1 = pd.DataFrame(results, columns = ['MODEL', 'Precision', 'Accuracy', 'Recall', 'F1-Score'])
# dtable1.sort_values(by=['Accuracy'], ascending=False)
dtable1
```
