#Task 6

```python
# Replace the missing values with the average of non-null values
# This function cleans the dataset
def clean(data):
  cols = ['RM']
  for col in cols:
    data[col].fillna(data[col].median(), inplace=True)
  return data
```

```python
data = clean(dt)
```
